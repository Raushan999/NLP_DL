# LSTM

So the problem with RNN was `vanishing/explodig` gradients
* It was not able to maintain the memory for the words occuring much past in sentence.
* So came the `LSTM`.

* LSTM solves the problem by keeping the important words in `long term` memory and keep updating it.
* Based on the `Short term` memory data.

* For that we keep another memory- `long term` apart from the general `short term` memory of a RNN.



* In LSTM, the thing inside the `box` is complex,
* That mechanism introduce `short term memory` and `long term memory` between each other

* In RNN, we just maintain a single state



* But in LSTM, we have the gates



1. Forget gate:- Remove the `long` term memory based on current input.
2. INput gate:- Add to the `long` term memory based on current input.
3. Output gate:- What from long term memory has to be sent to next states /output.

## summary

* Input:- `long-term` memory of previous state, `short-term` memory of previous state, and current word ($X_t$)



* Output:- $C_t$ long-term memory for next stage, $h_t$ short term memory for next stage.
  
 

* Inside the cell, we upadte the `long-term` memory ie., `add`,  `remove`.
* and `create` short term memory for the next stage.

# LSTM Architecture

* In order to maintain the `long-term` memory, concept of `gates` are very important here.
* There are `three` gates inside it:- `forget`, `input` and `output` gate.
* `forget` gate removes things from the `cell-state.`
* `input` gate adds things to the `cell state`.
* `output` gate calculate the `hidden` state for the next step.

# Deep RNN  || Stacked RNN || Stacked LSTM || Stacked GRUs

* multiple SimpleRNN  stacked together makes deep rnn.

* Here the `hidden state` $h_{t}$ information is passed to the next time step $h_{t+1}$ of current layer $H_l$.
* Also, $h_{t}$ of current layer $H_l$ is passed to the next layer $H_{l+1}$.
* So any node in the deeper layers receive output, ie, $h_t$ from $h_{t-1}$ of its own layer and $h_{t-1}$ of previous layer.

## Why and When to use?

* Initial layers find the word-wise sentiments.
* Middle layers find meaning in sentence.
* And final layers find the meaning in overall text.

* For encoder-decoder.


* For complex task (speech recognition)
* Large dataset.

# Bidirectional RNNs

#### when to use:-
* The past output depend on future input
* In this case normal RNNs/LSTMs fail
* eg, Named entity recognition
* - I went to `Amazon`. It is great website
  - I went to `Amazon`. It is a great river.
* Here website/river is affecting the tag of `amazon`.

* It has capability of understnading in both the direction.



Here to calculate $y_1$, the input $X_{11}$ is used and the future inputs $X_{12}$,$X_{13}$,$X_{14}$ ie., `the`, `best`, `website` are also used.

We concate the two outputs from both the sides.

and thus $y_1$ will be tagged as a `website` / organization



We are just using a wrapper of Bidirectional on the simple RNNs.

Only the no of weigths will be doubled as two RNNs are being used.
