# Transfer Learning

#### Problem with training of own model:
1. Deep learning models are data hungry, need lots of data.
2. Too much time to train them on our custom dataset.

#### Problem with using `pre-trained` models like `ImageNet`
* That model might not be trained on the classes our dataset has.
* 

`transfer learning`:- the reuse of a pre-trained model on a new problem

eg in real life:-
* if you know bicycle, you'll easily ride bike
* knowing to code in c++, helps you code in java

## Working of transfer learning



In this image of vgg model, conv layers are there to capture the spatial information

Fully connected part bascially help in classification.

What if our use case is not about classifying `cats` and `dogs`.

In this case, we cut the `conv` layers and freeze it.



Thereafter, we add our own fully connected layers, and the output layers.


`freezing` the conv layer, helps to `retain the weights values`, and restrict the re-training of that part of the new model.

## Why does transfer learning work?

* So many `conv` layers are build to extract different features of image.
* Initial features decode `primitive` features, which are present in almost all the use cases
* So the idea is ```Dont' re-invent the the wheel```, make vehicle out of it.

# ways of doing transfer learning

1. Feature extraction
2. `Fine tuning`

Feature extraction:- when the model is already trained on the use-case we are trying to solve. (eg. for dog/cat classification, we can use imagenet models)

Fine tuning:-  When the use case is completely different. Then we also change the `last few convolution layers` apart from the `fully connected layers`.
