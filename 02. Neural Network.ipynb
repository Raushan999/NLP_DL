# Perceptron

* An algorithm - used for supervised learning.
* A simple type of artifician neural network algorithm developed in 1957.
* Basic unit of a neural network, taking multiple binary inputs and producing a single binary output.
* It computes the weighted sum of its input, applies an activation function, and produces an output.

Design of a perceptron: a mathematical function.

> 
>
> * input --> weights + biases --> Z- value ---> activation function --> output

#### Neuron V/s Perceptorn


* Difference between neuron and perceptron:
* - Complex
* - Processing ( we don't what's goin in neuron)
* - Neuroplasticity ( neuron evolve with time )

# perceptron demo:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("placement.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
# plotting the points:
sns.scatterplot(data = df, x = 'cgpa', y = 'resume_score',hue= 'placed')
```

```python
# importing the perceptron model algorithm; it will try to find the weight and bias corresponding to
# the  input data and fit the best line 
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:, -1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X, y) # training the perceptron ( linear model) based on the data.
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X.values, y.values, clf = p, legend =2)
```
