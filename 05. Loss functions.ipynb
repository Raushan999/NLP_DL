# Loss Functions

* We can't improve anything if we can't measure it.
* So we need to find the loss, and then decide what to do.

Steps of applying loss:

* Give the first input (cgpa, iq) to the neural network, it will calculate the weights and biases of different layers.
* Then based on them output will be calculated.
* Based on the ouput, we find the deviation from the actual target. o/p - actual  = loss.
* Then we try to minimize the loss by `adjusting` the weights and bias using optimization techiniques (back propagation)
