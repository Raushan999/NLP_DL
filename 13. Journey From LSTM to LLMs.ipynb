# 1.Encoder - Decoder

 




###  Working mechanism:
* - Input --> Compressed by `encoder` --> Given to `Decoder` ---> O/p
  - Whole input is transformed to `context-vector` inside the `state` and then passed to `decoder`.
* So the whole load of `compressing` the input to `context-vector` is with the `state`.
* And so for longer texts, the retaintion of initial text was not maintained properly, `memory-loss`.
* Final output completely depended on what was there in the `context-vector`.

* `Failed to work on Large-dataset`

* So came the `Attention-Mechanism`:- 

# 2. Attention Mechanism

* Here the architecture was same as `Encoder-Decoder`, but for every output item, we have the access to all the input `hidden/cell` states.



### what was the problem with the earlier Encoder-Decoder:?

A potential issue with this encoder–decoder approach is that a neural network needs to be able to
compress all the necessary information of a `source sentence` into a `fixed-length vector`. 
This may make it difficult for the neural network to cope with `long sentences`, especially those that are longer
than the sentences in the training corpus. 



#### Performance of Attention-Mechanism approach :

* RNNsearch50 -> Length = 50
* RNNsearch30 --> lenght = 30
* RNNenc  -> Encoder architecture (earlier one)

We can see that performance `Bleu` score increased even on increasing the sentence length.

But for the traditional `enc-dec` architecture it deteriorated with increase in sentence length.



# Working of attention based Encoder-Decoder:

At any step of decoder, it has the access to all the encoders states (i;e, LSTMs $C_t$ and $h_t$)
Internal state is available everytime for the decoder.

* For ex:-
  
text = "my name is raushan".

Decoder has the access to all the states of encoder which are "my", "my name", "my name is",....


* So which hidden state is useful in predicting the translation of "my" ?
* Here comes the `Attention`, it finds the best possible state which will be useful for current decoder output.

\


* So we've all the states of encoder available to the decoder.
* And so the attention layer will try to find the best state for any current output.
* Once found, it will convert it to a `context-vector` and  give the output.





* How is it different from the normal `encoder-decoder` architecture?
* - There we'had only one `context-vector`, here for every output we've a context vector, and the initial output will derive more meaning from the initial input. Thus resolving the issue of `Long term memory loss`.

# Problem with the Attention mechanism / attention based encoder-decoder.

* Computational complexity due to mapping of each output to each encoder hidden states
* To find the context vector, similarity score was calculated for each output with all the inputs, and the best was used to find the context vector. $N^2$ complexity

* But later researchers realised that the problem was not in the attention mechanism, but in using `LSTM`, which work in sequential order.
* So they started working on eliminating the sequential operation and finding any `parallel processing`.
  
* So came `transformer`

# Transformers

* they eliminated `lstms` and introduced `attention` called `self-attention`



It still had `encoder-decoder` but use of `LSTMs` was discarded.

* Before `tranformers` the most bottleneck was that at one time you could only process only one input

* But in transformers, it can see all the inputs simultaneously.

Actual architecutre introduced in the paper:- 2017:- 



* everyting was parallelely computed

# Transfer Learning

* Till now transfer learning was used only in `vision`.
* But now, it was proposed first time 

Here the training is done on the existing `state-of-the-art` existing models
* Only we change the last layers/output and train according to our use-case
* It works, bcoz the previous layers have already learnt basic features like edges/shapes/lines/ etc., it can now easily classify b/w images

### Why till 2018, transfer learning was not used in NLP tasks ??

* Task specificity:
* - NER, Sentiment analysis, POS, ML, QA, summarization
  - All are different tasks.
  - And they thought a single model can't work on all  these tasks.
* So transfer learning was not seen in NLP till now.
* But it changed after `ulm-fit` research paper.

* In the paper they didn't use `machine translation` for pre-training.
* Rather they used, `language modelling`
* In language modlling, we predict the next word, like `next-word-prediction`.

### Why language modelling as a pretraining tasks got so much success?

#### 2 reasons: - 
1. Rich feature learning:- learning grammers, context, semantic etc while training.
2. So once it has already learnt these basic tasks, it can be used to other differet complex tasks - NER, summarization
3. Huge availability of data:- can take millions of data from internet and generate dataset without bothering about the labelling.
4. I.e., making it un-supervised.

#### Set up used in the paper: transfer learning:

AWD lSTM  --> Wikipedia ( unsupervised pretraining) --> language modelling --> Once trained --> output layer changed to a classifier  on imdb/ yelp data [ fine tuning]  --> got new model
