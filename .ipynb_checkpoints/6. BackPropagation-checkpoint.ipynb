# Backpropagation: What ?

* Backprop is just a very simple process to tell us ```which parameters to change in a neural network.```

* Backpropagation:- An Algorithm --> to train a NN --> (to find the optimal weights and bias)

 
 

# steps of backpropagation

* [0] Initialize the weights and bias ${W}$, ${b}$ :- random or ${W} = 1$ , ${b}  = 0$
* For $i$ in $range(total data)$:- (For $ N $ no of epochs)
  
* - [1] Select a row data (training data)
* - [2] Predict the output using := `forward propagation`
* - [3] Choose a `loss function`
* - [4] Move backward from the output layer:
   * - Update the weights of last weights connected to output layer
     - Update the next previous layer weights/bias
     - Move backward till all the weights and biases are updated for a given row data using GD
* Do it for all the rows

# Back Propagation: The Why?

## intution behind the algorithm:


* Loss function is a function of all the trainable parameters
* So while finding the gradient, we find it wrt to all the parameters i.e., `partial derivate` wrt all the parameters.

concept of derivate:

* $dy$/$dx$ is the rate of change in y wrt `one` unit change in `x`

concept of minima:
* we need to find the minima wrt to all the parameters in those many dimension

Effect of learning rate:
* it makes the movement in a updated direction smooth(no drastic change)

When to stop: criteria of convergence:
* When the update in weight becomes negligible, ie. derivate becomes `zero`.

## memoization:
* core technique used in dynamic programming

````python
# fibonacci implementation:
d = {0:1, 1:1}
def fib(n):
    if n in d:
        return d[n]
    else:
        d[n] = fib(n-1) + fib(n-2)
        return d[n]
print(fib(36))
````

# memoization in neural network:
* it is storing of the derivatives so that it can be used again for later derivatives

# Summary

* Backpropagation:- `chain rule multiplication` + `memoization of derivatives`

# Gradient Descent

### Batch gradient descent:
* all data points prediction done together using dot product
* then you will get an array of prediction
* then calculate the loss
* then update the weights

Weight updated:- no of epochs times

### Stochastic GD
* Take a random point
* calculate the loss
* update the weights for that point.

Weight updated:- `no of training points` * `no of epochs times`

* More time taking than Batch GD

### Which is faster:

* for a given epochs Batch is faster
* But SG finds the optimal weights faster, faster convergence for same given epochs

Benefits of SG:
* To move out of local minima
Limitation of SG:
* can come out of global minima, since it is too randomized.

Mini Batch GD:
* batch_size = 20, middle of two GD

Vectorization:
* replacement of for loop (lesser time complexity)
* but loading a larger dataset on to RAM is costly, so we use loop that time
