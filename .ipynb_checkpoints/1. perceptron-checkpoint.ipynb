# Perceptron trick: How to train a perceptron.

It computes the `weighted sum` of its input, applies an `activation function`, and produces an output.

```python
# pip install scikit-learn
# !pip install matplotlib
```

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples = 100, n_features=2, n_informative=1, n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,
                           class_sep=10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10,4))
plt.scatter(X[:,0], X[:,1], c= y)
```

```python
# function to find the weights and bias using our perceptron algorithm.
def perceptron(X, y):
    X = np.insert(X, 0, 1, axis = 1)
    weights = np.ones(X.shape[1]) # X.shape = (100,2)
    lr = 0.01
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j], weights))
        weights = weights + lr*(y[j] - y_hat)*X[j]
        
    return weights[0], weights[1:]
    
# step function:
def step(a):
    return a > 0 # if dot product is +ve return 1 else 0.
```

```python
# intercept and the weights:
intercept_, coef_ = perceptron(X, y)
```

```python
intercept_, coef_
# Ax + By + C = 0 ---> will give y = mx + b --> m = -A/B, b = -C/B
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
m, b
```

```python
# now we have m and b, we can plot this line.
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
# plotting the line on the dataset.
plt.figure(figsize=(6,3))
plt.plot(x_input, y_input, color = 'red', linewidth = 3)
plt.scatter(X[:,0], X[:,1], c = y, s = 100)
plt.ylim(-3,2)
```

```python
# again finding the weights and bias using loss function:

X, y = make_classification(n_samples = 100, n_features=2, n_informative=1, n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,
                           class_sep=10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10,4))
plt.scatter(X[:,0], X[:,1], c= y)
```

```python
# function to find the weights and bias using our perceptron algorithm.
def perceptron(X, y):
    w1 = w2 = b = 1
    
    lr = 0.01
    for i in range(1000):
        for j in range(X.shape[0]):
            # check condition
            z = w1*X[j][0] + w2*X[j][1] + b
            if z*y[j] < 0: # otw max(_ , _ ) will return 0.
                w1 = w1 + lr* y[j]*X[j][0]
                w2 = w2 + lr* y[j]*X[j][1]
                b = b + lr* y[j]
        
        
    return w1, w2, b
    
```

```python
w1, w2 , b = perceptron(X, y)
```
