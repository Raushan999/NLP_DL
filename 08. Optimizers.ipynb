## Momentum optimization: why?




Intution:-
* If most of the previous `gradients` are saying to move in a particular direction, you shoud `increase` your speed.
* You can trust on your neighbors and `accelerate` your search.

Impotance of using:

## Momentum : Mathematical::
* It uses exponential moving avg.
  



#### Mathematical intution of momentum:-

* We are getting push from `2` components:
* - The gradient at that point
  - The preivous velocity 
* For $\beta$ = 0  -> it will be simply `stochastic grad descent`
* For $\beta$ = 1 --> no decay will be seen. not desired.


Benefit:
* Has enough speed to `come out of local minima`.


Disadvantage:-
* `Oscillation`
* althogh it come out of local minima, but at global minima it starts oscillating before settling down, time consuming.

 

### Nesterov Accelerated Gradient: NAG

* it is same as Momentum, only a simple change in the way of jump.

Here first we take the jump based on momentum, and then calulate the `gradient` at that newly jumped point.

* So the gradient at new point helps us to move in right direction to `minima` and the momentum helps us with `speed`. Thus `speed` + `delayed gradient` helps in preventing `oscillation`.
* But for the simple momentum SGD, when we are around `minima`, the `speed` and `gradient` add and become larger, thus leading to take larger jump to other side of minima. From there again, we try returning..

* In NAG, we know what will happen ahead, and so we take the steps.

Disadvantages:-

* We are `damping` the oscillation, that can lead to `local minima` case.

# AdaGrad:- Adaptive Gadient

when to use:-
* When the features are spase, ie. data mostly `0`. [0,10,0,0,0,1,1,10,0,9,0,0,0,0,]
* Sparsity creates `elongated bowl problem`. so along one dimension, contour plot get elongated.



* So with this `sparse` data, when we apply, SGD or Momentum, the update in sparse data dimension is lesser as compared to non-sparse.
* Since we are using the same learning rate.

Mathematical intution of sparsity:-

Let's take `x1` as sparse and `x2` feature as normal.

* In Batch GD, loss is calculated after a whole batch, so the total loss sum calculated for `x1` will $<<<$ than `x2`.

* Hence the convergence will be slower in `x1` direction.

Adagrad Mathematical intution:

Here we keep different `lr` for the two features.



* For sparse data, $V_t$ is less so the `lr` increases, and thus makes the weight update faster.
* For non-sparse data, `lr` is slowed down.


Disadvantages:-

* Never reaches to Optimal solution, global solution
* Since the `lr` keeps decreasing as $ V_t$ keeps increasing
* But this concept will be used later in `adam`

# RMS prop:- Root Mean Square Prop

* Improvement over `adagrad`





* Here we are not taking the whole history of gradient
* we are applying `EWMA`  and providing more weightage to recent grads
* Thus the $V_t$ is not increasing too much
* Hence, the `lr` is not going too small.

Here, $V_t$ for higher epochs are getting more weightage, than the past epochs, thus keeping the `lr` checked.
