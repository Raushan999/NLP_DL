1. Long term dependencies:- `Vanishing Gradient`

 

so the gradients calculated for this goes to `0`. Only short term dependencies are valued while calculating the derivative of loss

Bcoz the deriavte of `tanh` - activation function lies in `(0-1)`, each time multiplying for 100+ times goes to `0`

#### Solution of vanishing grads:= 

1. Different activation - relu/ leakyRelu
2. better weight init
3. Skip rnns
4. LSTM

### Problem 2: Exploding grads

* derivatives too large: if `relu` taken and if value is `1` then `1` multiplied with other values (`Xi`), lead to bigger numbers.
* Learning rate too high.

Solutions to Exploding grads:- 
1. Gradient clipping:- setting grad to a max limit. not greater than this
2. controlled learning rate
3. LSTM
